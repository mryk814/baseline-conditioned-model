# Compare Baseline-Conditioned Models

Run this notebook after installing `requirements.txt`. It mirrors the Streamlit demo's model comparison path.

```python
from src.generate_data import generate_material_pairs
from src.evaluation import split_random, evaluate_models
from src.models import GlobalAbsoluteLinearModel, GlobalDeltaLinearModel
from src.local_linear import LocalLinearRegressor, PartialPoolingLocalLinearRegressor
from src.gp_models import GPAbsoluteModel, GPDeltaModel

data = generate_material_pairs(random_state=42)
train, test = split_random(data, random_state=42)
models = {
    'absolute_linear': GlobalAbsoluteLinearModel().fit(train),
    'delta_linear': GlobalDeltaLinearModel().fit(train),
    'local_linear': LocalLinearRegressor().fit(train),
    'partial_pooling': PartialPoolingLocalLinearRegressor().fit(train),
    'gp_absolute': GPAbsoluteModel(max_train_size=260).fit(train),
    'gp_delta': GPDeltaModel(max_train_size=260).fit(train),
}
evaluate_models(models, test)
```
